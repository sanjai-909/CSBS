```python
from collections import defaultdict

# Stores the production rules
productions = defaultdict(set)
# Stores the FIRST and FOLLOW sets
firstsets = defaultdict(set)
followsets = defaultdict(set)

# Global variable to store the starting symbol
FIRST = ''

def findfirst(c):
    if not c.isupper():
        firstsets[c].add(c)
        return
    
    for rhs in productions[c]:
        if rhs[0] == '#':
            firstsets[c].add('#')
        else:
            for ch in rhs:
                if not ch.isupper():
                    firstsets[c].add(ch)
                    break
                else:
                    findfirst(ch)
                    firstsets[c].update(firstsets[ch])
                    if '#' not in firstsets[ch]:
                        break

def findfollow(c):
    if FIRST == c:
        followsets[c].add('$')
    
    for lhs, rhs_set in productions.items():
        for rhs in rhs_set:
            for i in range(len(rhs)):
                if rhs[i] == c:
                    if i + 1 < len(rhs):
                        findfollowfirst(rhs[i + 1], c)
                    if i + 1 == len(rhs) or ('#' in firstsets[rhs[i + 1]]):
                        if c != lhs:
                            findfollow(lhs)
                            followsets[c].update(followsets[lhs])

def findfollowfirst(c, lhs):
    if not c.isupper():
        followsets[lhs].add(c)
    else:
        for ch in firstsets[c]:
            if ch != '#':
                followsets[lhs].add(ch)
            else:
                followsets[lhs].update(followsets[c])
                findfollow(c)

def add_productions():
    global FIRST
    productions['E'] = {"TA"}
    FIRST = list(productions.keys())[0]

    productions['A'] = {"+TA", "#"}
    productions['T'] = {"FB"}
    productions['B'] = {"*FB", "#"}
    productions['F'] = {"(E)", "id"}

def main():
    add_productions()
    done = set()

    # Calculate FIRST sets
    for c in productions:
        if c not in done:
            findfirst(c)
            done.add(c)

            print(f"First ({c}) = {{", end="")
            for ch in sorted(firstsets[c]):
                print(f"{ch},", end="")
            print("}")

    print("\n\n---------------------------------------------------------------\n\n")
    
    done.clear()
    
    # Calculate FOLLOW sets
    for c in productions:
        if c not in done:
            findfollow(c)
            done.add(c)

            print(f"Follow ({c}) = {{", end="")
            for ch in sorted(followsets[c]):
                print(f"{ch},", end="")
            print("}")

if __name__ == "__main__":
    main()
```
