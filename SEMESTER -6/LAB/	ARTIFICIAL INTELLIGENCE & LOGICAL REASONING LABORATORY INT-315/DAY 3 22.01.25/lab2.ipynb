```python
def get_graph_from_user():
    graph = {}
    num_nodes = int(input("Enter the number of nodes: "))

    for _ in range(num_nodes):
        node = input("Enter node name: ")
        neighbors = input(f"Enter neighbors of {node} (comma-separated): ").split(",")
        graph[node] = [neighbor.strip() for neighbor in neighbors if neighbor.strip()]

    return graph

def find_all_paths(graph, start, end, path=None):
    if path is None:
        path = []
    path = path + [start]
    
    if start == end:
        return [path]
    if start not in graph:
        return []
    
    paths = []
    for node in graph[start]:
        if node not in path:
            new_paths = find_all_paths(graph, node, end, path)  
            for p in new_paths:
                paths.append(p)
    
    return paths



graph = get_graph_from_user()

source = input("Enter the source node: ")
destination = input("Enter the destination node: ")

all_paths = find_all_paths(graph, source, destination)
print("\nAll possible paths from {} to {}:".format(source, destination))
for path in all_paths:
    print(" -> ".join(path))
```

```python
from itertools import permutations

kids = [4, -2, 2]
chocolates = [5, 4, 5]

chocolate_permutations = list(permutations(chocolates))

results = []

for perm in chocolate_permutations:
    differences = [abs(kids[i] - perm[i]) for i in range(len(kids))]
    max_difference = max(differences)
    results.append((list(perm), differences, max_difference))

for i, (assignment, differences, max_diff) in enumerate(results, 1):
    print(f"Possible way {i}:")
    print(f"Kids:       {kids}")
    print(f"Chocolates: {assignment}")
    print(f"Differences: {differences}")
    print(f"Max Difference: {max_diff}\n")
```

```python
# 1.b All path problem  -user input
def all_paths(graph, start, end, path=None):
    """
    Return a list of all simple paths from 'start' to 'end' in 'graph'.
    """
    if path is None:
        path = []
    path = path + [start]  # Create a new path list with 'start' appended
   
    # If start == end, we have found a complete path
    if start == end:
        return [path]
   
    # If the start node isn't in the graph, return no paths
    if start not in graph:
        return []
   
    paths = []
    for neighbor in graph[start]:
        if neighbor not in path:
            new_paths = all_paths(graph, neighbor, end, path)
            # Extend the 'paths' list by all new paths found
            for np in new_paths:
                paths.append(np)
    return paths


# --------------------- MAIN PROGRAM ---------------------
def main():
    # 1. Ask user how many vertices
    n = int(input("Enter the number of vertices: "))
   
    # 2. Build the graph by reading adjacency lists
    graph = {}
    for i in range(n):
        # Example: user might enter 'A' for the first vertex
        vertex = input(f"Enter the name of vertex {i+1}: ")
       
        adjacency = []
        print(f"Enter neighbors of vertex '{vertex}' one by one.")
        print("Type 'no' when you're done adding neighbors.\n")
       
        while True:
            neighbor = input(f"   Neighbor of {vertex}: ")
            if neighbor.lower() == 'no':
                break
            adjacency.append(neighbor)
       
        graph[vertex] = adjacency
   
    # 3. Ask user for start and end nodes
    start_node = input("Enter start node: ")
    end_node   = input("Enter goal (end) node: ")

    # 4. Find all paths
    result_paths = all_paths(graph, start_node, end_node)

    # 5. Print results
    print(f"\nAll paths from {start_node} to {end_node}:")
    for p in result_paths:
        print(" -> ".join(p))


# Run the main function
if __name__ == "__main__":
    main()
```
